# Sentimentanalyse vertiefen: Märchen vergleichen


```python
# Hilfsfunktionen laden
import sys
sys.path.append('..')
from kurs_helpers import (
    lade_bereinigten_text,
    GERMAN_STOPWORDS,
    GRIMM_SENTIMENT_LEXIKON,
    berechne_sentiment,
    berechne_sentiment_mit_negation,
    pruefe_08_aufgabe_1,
)
```

Im vorherigen Kapitel haben wir den Sentiment Score für ein einzelnes Märchen berechnet. Jetzt wollen wir einen Schritt weitergehen und die Stimmung von **zwei Märchen miteinander vergleichen**: "Schneewittchen" und "Hänsel und Gretel".

Die spannende Frage: Welches der beiden Märchen ist "düsterer"? Und stimmt unser Bauchgefühl mit dem überein, was die Analyse sagt?

## Schritt 1: Beide Märchen laden und vorbereiten


```python
# Schneewittchen laden und bereinigen
bereinigter_text_schneewittchen, wort_liste_schneewittchen = lade_bereinigten_text('Sneewittchen.txt')

# Stoppwörter und Zahlen entfernen
schneewittchen_gefiltert = []
for wort in wort_liste_schneewittchen:
    if wort not in GERMAN_STOPWORDS and not wort.isdigit():
        schneewittchen_gefiltert.append(wort)

print("Schneewittchen: ", len(wort_liste_schneewittchen), "Wörter gesamt: ", len(schneewittchen_gefiltert), "nach Filterung")
```

```python
# Hänsel und Gretel laden und bereinigen
bereinigter_text_haensel, wort_liste_haensel = lade_bereinigten_text('HaenselGrethel.txt')

# Stoppwörter und Zahlen entfernen
haensel_gefiltert = []
for wort in wort_liste_haensel:
    if wort not in GERMAN_STOPWORDS and not wort.isdigit():
        haensel_gefiltert.append(wort)

print("Hänsel und Gretel: ", len(wort_liste_haensel), " Wörter gesamt, ", len(haensel_gefiltert), "nach Filterung")
```

## Schritt 2: Sentiment Score berechnen

Jetzt berechnen wir den Sentiment Score für beide Märchen — einmal ohne und einmal mit Negationserkennung.


```python
# Schneewittchen: einfacher Score
score_sw_einfach, woerter_sw_einfach = berechne_sentiment(schneewittchen_gefiltert)
print("Schneewittchen (ohne Negation): Score =", score_sw_einfach, ", gefundene Wörter =", len(woerter_sw_einfach))

# Schneewittchen: Score mit Negationserkennung
score_sw_negation, details_sw = berechne_sentiment_mit_negation(schneewittchen_gefiltert)
print("Schneewittchen (mit Negation):  Score =", score_sw_negation)
```

```python
# Hänsel und Gretel: einfacher Score
score_hg_einfach, woerter_hg_einfach = berechne_sentiment(haensel_gefiltert)
print("Hänsel und Gretel (ohne Negation): Score =", score_hg_einfach, ", gefundene Wörter =", len(woerter_hg_einfach))

# Hänsel und Gretel: Score mit Negationserkennung
score_hg_negation, details_hg = berechne_sentiment_mit_negation(haensel_gefiltert)
print("Hänsel und Gretel (mit Negation):  Score =", score_hg_negation)
```

## Schritt 3: Ergebnisse vergleichen

Schauen wir uns die Ergebnisse nebeneinander an:


```python
print("--- Ergebnisse ---")
print()
print("SCHNEEWITTCHEN:")
print("  Wörter gesamt:", len(wort_liste_schneewittchen))
print("  Wörter gefiltert:", len(schneewittchen_gefiltert))
print("  Emotionale Wörter:", len(woerter_sw_einfach))
print("  Score (ohne Negation):", score_sw_einfach)
print("  Score (mit Negation):", score_sw_negation)
print()
print("HÄNSEL UND GRETEL:")
print("  Wörter gesamt:", len(wort_liste_haensel))
print("  Wörter gefiltert:", len(haensel_gefiltert))
print("  Emotionale Wörter:", len(woerter_hg_einfach))
print("  Score (ohne Negation):", score_hg_einfach)
print("  Score (mit Negation):", score_hg_negation)
```

## Schritt 4: Detailanalyse

Welche positiven und negativen Wörter stecken in den Märchen?


```python
# Schneewittchen: positive und negative Wörter
positive_sw = []
negative_sw = []
for wort in woerter_sw_einfach:
    if GRIMM_SENTIMENT_LEXIKON.get(wort, 0) > 0:
        positive_sw.append(wort)
    elif GRIMM_SENTIMENT_LEXIKON.get(wort, 0) < 0:
        negative_sw.append(wort)

print("SCHNEEWITTCHEN")
print("  Positive Wörter (", len(positive_sw), "):", positive_sw)
print("  Negative Wörter (", len(negative_sw), "):", negative_sw)
```

```python
# Hänsel und Gretel: positive und negative Wörter
positive_hg = []
negative_hg = []
for wort in woerter_hg_einfach:
    if GRIMM_SENTIMENT_LEXIKON.get(wort, 0) > 0:
        positive_hg.append(wort)
    elif GRIMM_SENTIMENT_LEXIKON.get(wort, 0) < 0:
        negative_hg.append(wort)

print("HÄNSEL UND GRETEL")
print("  Positive Wörter (", len(positive_hg), "):", positive_hg)
print("  Negative Wörter (", len(negative_hg), "):", negative_hg)
```

## Schritt 5: Negation im Detail

Schauen wir uns an, wo die Negationserkennung eingegriffen hat:


```python
negierte_sw = []
for wort, orig, negiert, end in details_sw:
    if negiert:
        negierte_sw.append((wort, orig, end))

negierte_hg = []
for wort, orig, negiert, end in details_hg:
    if negiert:
        negierte_hg.append((wort, orig, end))
print("Negierte Wörter in SCHNEEWITTCHEN:")

if negierte_sw:
    for wort, orig, end in negierte_sw:
        print("  '" + wort + "':", orig, "→", end)
else:
    print("  Keine Negationen gefunden.")

print()
print("Negierte Wörter in HÄNSEL UND GRETEL:")

if negierte_hg:
    for wort, orig, end in negierte_hg:
        print("  '" + wort + "':", orig, "→", end)
else:
    print("  Keine Negationen gefunden.")
```
