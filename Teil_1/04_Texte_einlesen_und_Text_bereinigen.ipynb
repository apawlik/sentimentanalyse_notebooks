# Texte einlesen und bereinigen


Wenn wir uns innerhalb der Geisteswissenschaften mit Python bewegen, ist die Arbeit mit Texten sehr relevant. Mit Python können wir Texte speichern, verändern und sogar ganze Textdateien einlesen und auswerten lassen.

Beginnen wir mit der Bearbeitung von Texten. Wie wir bereits wissen, können Texte bzw. Strings in Variablen gespeichert werden. Nehmen wir bspw. den Titel des Werks "Die Sterntaler":


````python
titel = "Die Sterntaler"
print(titel)
````

Wir können den Text auch verändern und normalisieren.

---

### Hinweis

Normalisieren bedeutet in diesem Kontext, dass ein Text vereinheitlicht wird, damit das Programm gleiche Inhalte als solche identifizieren kann. Ein Schritt der Normalisierung ist die sog. "Case Conversion", bei der der gesamte Text in Kleinbuchstaben überführt wird. Denn wie wir bereits gelernt haben, grenzt Python nach Groß- und Kleinschreibung ab ("Märchen" und "märchen" wären zwei unterschiedliche Werte).

---

Die Kleinschreibung können wir zunächst mit `.lower()` erreichen:


````python
print(titel.lower())
````

**Aber Achtung:** die kleingeschriebene Variante ist nicht in der Variablen gespeichert. Sie wurde nur einmalig ausgegeben.

## Text einlesen

Nachdem wir den Titel nun für Python normalisiert haben, wollen wir den Text des Märchens einlesen. Aber wie machen wir das? Wir müssen zum Glück nicht den gesamten Märchentext in einen String schreiben und speichern. Denn wir können ganze (Text-)Dateien einlesen, in einer Variablen speichern und dann später verändern bzw. auswerten. Hierzu nutzen wir in einem ersten Schritt die Funktion `with open('Dateipfad zum Text', 'r', encoding='utf-8') as datei:` und in einem zweiten die konkrete Anweisung zum Lesen der Datei mit `sterntaler = datei.read()`.

*Hinweis:* Das `'r'` steht für "read", also, dass die Datei gelesen wird aber die Originaldatei nicht von uns geändert oder überschrieben wird. Das Encoding nach utf-8 hilft, dass Sonderzeichen bspw. richtig angezeigt werden.

Wir nehmen das Märchen "Die Sternthaler":


````python
with open('../maerchen_texte/Sternthaler.txt', 'r', encoding='utf-8') as datei:
    sterntaler = datei.read()
````

Nun ist der gesamte Text in der Variablen *sterntaler* gespeichert. Wir können uns anzeigen lassen, wie lang der Text ist, also die Anzahl der Zeichen ausgeben lassen:


````python
print(len(sterntaler))
````

Wir haben nun die Information erhalten, dass das Märchen "Die Sternthaler" 1587 Zeichen enthält. Gemeint sind damit keine Wörter, sondern tatsächlich einzelne Zeichen: Buchstaben, Satzzeichen, womöglich auch Zahlen.

## Text bereinigen

Um valide mit dem Text arbeiten zu können, müssen wir auch im Text für Einheit und Vergleichbarkeit sorgen. So, wie wir den Titel des Märchens normalisiert haben, können bzw. müssen wir das auch für den gesamten Text tun.
Damit Wörter vergleichbar sind und Satzzeichen nicht stören, können wir mit einfachen Methoden alles klein schreiben, die Punkte und Kommata durch "Nichts" ersetzen (bzw entfernen) und dann den Text in eine Liste von Wörtern zerlegen, wobei wir am Leerzeichen eine Trennung vornehmen.

Für die Kleinschreibung aller Wörter im Text nutzen wir die eingebaute Funktion `lower()`. Diese haben wir zuvor auf den Titel angewendet und tun das nun für den gesamten Text, den wir in der Variablen *sterntaler* gespeichert haben, und speichern ihn in der Variablen *sterntaler_klein*:


````python
sterntaler_klein = sterntaler.lower()
````

Auch Satzzeichen brauchen wir für unser Vorhaben nicht. Diese können wir ebenfalls mit einer von Python vorgegebenen Funktion direkt entfernen: `replace()`

**Wichtig ist:** Wir bestimmen mit der Funktion, worauf sie angewendet werden soll und was durch etwas anderes ersetzt wird. 

Hier sagen wir Python also: Nimm den kleingeschriebenen Text (*sterntaler_klein*) und ersetze (*replace*) den Punkt `"."` mit `""` Nichts (Anführungszeichen ohne Inhalt). Dieses Vorgehen kann ich für die weiteren Satzzeichen beliebig oft wiederholen, bis alle Satzzeichen durch Nichts ersetzt sind (im nachfolgenden Beispiel wird zweimal `replace()` angewendet). Den so entstandenen neuen Text ohne Satzzeichen speichere ich dann in der Variablen *bereinigter_text*:


````python
bereinigter_text = sterntaler_klein.replace(".", "").replace(",", "")
````

Nun zerlegen wir den gesamten Text in einzelne Wörter und speichern diese einfach in einer Liste. Die Liste nennen wir `wort_liste`. Dafür wenden wir wieder eine eingebaute Funktion an: `split()` und lassen uns zur Kontrolle die ersten 10 Listenelemente ausgeben:

`split()` zerlegt eine Zeichenkette in einzelne Elemente und trennt am Leerzeichen.

*Hinweis:* Man kann selbst bestimmen, an welchem Zeichen getrennt wird. Mehr Informationen: https://www.w3schools.com/python/ref_string_split.asp


````python
wort_liste = bereinigter_text.split()
# gib die ersten 10 Wörter aus:
print(wort_liste[:10])
````

## Wörter zählen

Wenn wir wissen möchten, aus wie vielen Wörtern (nicht Zeichen!) unser Text bzw. unsere Liste besteht, wäre es mühsam, diese manuell zu zählen. Glücklicherweise bietet Python die Möglichkeit, solche Prozesse zu automatisieren – mithilfe vorgefertigter Programmpakete, sogenannter Bibliotheken.“

### Bibliotheken

Wörter sind nützlich, aber noch nützlicher sind die Sätze und Geschichten, die wir mit ihnen bauen. Ähnlich verhält es sich mit Python: Während viele leistungsstarke, allgemeine Werkzeuge in Python eingebaut sind, leben spezialisierte Werkzeuge, die aus diesen Grundeinheiten aufgebaut sind, in Bibliotheken, die bei Bedarf aufgerufen werden können.

Eine Python-Bibliothek ist eine Sammlung von Dateien (genannt Module), die Funktionen enthält, die Sie in Ihren Programmen verwenden können. Einige Bibliotheken (auch als "Pakete" bezeichnet) enthalten Standarddatenwerte oder Sprachressourcen, auf die Sie in Ihrem Code verweisen können. Bisher haben wir die Python-Standardbibliothek verwendet, die eine umfangreiche Sammlung von eingebauten Modulen enthält. Einige beliebte Bibliotheken sind:

- **Pandas** - Werkzeug zur tabellarischen Datenanalyse
- **Matplotlib** - Werkzeuge zur Datenvisualisierung
- **BeautifulSoup** - für das Parsen von HTML- und XML-Dokumenten
- **Requests** - für HTTP-Anfragen (z. B. für Web-Scraping, Verwendung von APIs)
- **Scikit-learn** - Werkzeuge zum maschinellen Lernen für die prädiktive Datenanalyse
- **NumPy** - Werkzeuge für numerische Berechnungen wie mathematische Funktionen und Zufallszahlengeneratoren

#### Import von Bibliotheken

Verwenden Sie `import`, um eine Bibliothek in den Speicher eines Programms zu laden. Dann können Sie auf Dinge aus der Bibliothek als `library_name.function` verweisen. Wir importieren und verwenden die String-Bibliothek, um eine Liste von ASCII-Kleinbuchstaben zu erzeugen und die Groß- und Kleinschreibung einer Textzeichenfolge zu ändern:

---

### Hinweis

ASCII steht für "American Standard Code for Information Interchange". Dahinter verbirgt sich ein Standard für Zeichensätze, durch den Zeichen (Buchstaben, Satzzeichen, Zahlen und Steuerzeichen) in eine einheitliche, maschinenlesbare Form gebracht werden.

---


````python
import string

# gibt die ASCII-Kleinbuchstaben aus
print(f'The lower ascii letters are {string.ascii_lowercase}')

# schreibt jeden ersten Buchstaben eines Wortes groß
print(string.capwords('capitalise this sentence please.'))
````

---

### Hinweis: Punktnotation

Im Zusammenhang mit der Methode `.append` haben wir die Python-Punktnotation, also die Schreibweise von Methoden durch einen Punkt, bereits kennengelernt. Wir können die gleiche Syntax verwenden, wenn wir Funktionen einer bestimmten Python-Bibliothek aufrufen, wie z. B. `string.capwords()`. Die Punktschreibweise ist in Python weit verbreitet und kann sich auf Beziehungen zwischen verschiedenen Typen von Python-Objekten beziehen. Denken Sie daran, dass das Objekt rechts vom Punkt immer ein Teil des größeren Objekts links ist.

## Verwende Aliase

Sie können `import ... as ...` verwenden, um einer Bibliothek beim Importieren einen kurzen Alias zu geben. Auf diese Weise müssen Sie nicht immer den vollen Namen der importierten Bibliothek verwenden, sondern können effizienter auf Elemente verweisen. Da pandas eine der populärsten Bibliotheken für Python ist, probieren wir das für diese Bibliothek aus:


````python
import pandas as pd
````

Wir können nun immer, wenn wir uns auf die Bibliothek pandas beziehen wollen, die Abkürzung `pd` verwenden.

Viele populäre Bibliotheken haben gemeinsame Aliasnamen. Zum Beispiel:

```python

```

Die Verwendung dieser gemeinsamen Aliasse kann die Arbeit mit vorhandener Dokumentation und mit Tutorials erleichtern.

Um die Wörter in unserer Liste zu zählen, brauchen wir eine der Bibliotheken, die Python zur Verfügung stellt. Diese müssen wir erst im Script importieren und dann können wir direkt mit der Funktion `Counter()` die Wörter in unserer Liste zählen.


````python
from collections import Counter

# Der Counter erledigt das Zählen für uns
# wir speichern die gezählten Wörter in unserer Wortliste in der Variablen haeufigkeiten
haeufigkeiten = Counter(wort_liste)
````

Wir können uns auch über die Funktion `most_common()` die häufigsten gezählten Wörter anzeigen lassen:


````python
# Die 10 häufigsten Wörter ausgeben die in haeufigkeiten gespeichert sind
print(haeufigkeiten.most_common(10))
````
