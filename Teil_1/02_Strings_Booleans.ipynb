## Slice

Um mehrere, aber nicht alle, Zeichen einer Zeichenkette abzurufen, können wir die Funktion "Slice" (Schneiden) verwenden.
Ein Slice ist ein Teil einer Zeichenkette, auf den wir mit `[start:stop]` verweisen können. Dabei ist "start" der Index des **ersten** gewünschten Zeichens und "stop" der Index des **letzten** Zeichens. Durch den Verweis auf ein String-Slice wird der Inhalt des ursprünglichen Strings nicht verändert. Stattdessen gibt das Slice eine Kopie des gewünschten Teils der ursprünglichen Zeichenkette zurück.


```python
title[0:3]
```

Beachten Sie, dass im obigen Beispiel `title[0:3]` mit Null beginnt, was sich auf das erste Element in der Zeichenkette bezieht, und mit einer 3 endet. Bei der Arbeit mit Slices wird der Endpunkt so interpretiert, dass er bis zur angegebenen Indexnummer reicht, diese aber nicht einschließt. Mit anderen Worten: Das Zeichen an der Indexposition 3 in der Zeichenkette Froschkönig ist "s", sodass das Slice [0:3] bis zu diesem Zeichen reicht, es aber **nicht einschließt**. Daher wird uns als Ausgabe "Fro" geliefert.

### Die Funktion len()

Die in Python eingebaute Funktion `len()` gibt die Länge eines Elements an. Im Falle einer Zeichenkette gibt sie also an, wie viele Zeichen die Zeichenkette hat.


```python
len('Grimm')
```

**Wichtig:** Im Gegensatz zum Index, der von 0-4 reichen würde, wird mit dieser Funktion die tatsächliche Anzahl der Zeichen einer Variable ausgegeben. Da 'Grimm' fünf Zeichen enthält, ist die Ausgabe von `len()` 5.

## Booleans

Booleans ist ein Datentyp, der nur zwei Werte annehmen kann: `True` oder `False`, also "Wahr" oder "Falsch". Sie repräsentieren also den Wahrheitsgehalt einer Aussage. Wie für Integers, Floats und Strings kannst du auch den Datentyp eines Booleans mit der Funktion `type()` überprüfen:


```python
x = True
y = False
print(type(x))
print(type(y))
```

Wir können auch einen Boolean relativ zu einer anderen Variablen definieren. Im folgenden Codeblock überprüfen wir, ob x größer als 3 ist:


```python
x = 5
y = x > 3
print(y)
```

Da wir x zunächst als 5 definiert haben und dieser Wert größer als 3 ist, wird uns `True`, also "Wahr" ausgegeben.
