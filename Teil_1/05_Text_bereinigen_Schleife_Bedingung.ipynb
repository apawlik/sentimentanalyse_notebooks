# Schleifen und Bedingungen


Nachdem wir unseren Text nun hinsichtlich der Groß- und Kleinschreibung sowie der Zeichensetzung normalisiert und bereinigt haben, bleibt noch ein Faktor, der unsere nachfolgende Analyse beeinflussen kann: Zahlen. Je nach Text befinden sich Zahlen, wie z. B. die Nummer des Märchens, innerhalb des Fließtextes. Um gut mit dem Text weiterarbeiten zu können, sollten wir diese entfernen. Hierzu benötigen wir sogenannte Schleifen und Bedingungen.

## Schleifen

Schleifen benötige ich, um dieselbe Aktion automatisiert mehrmals durchzuführen. Wir haben beispielsweise eine Liste von Märchenfiguren:


```python
figuren = ["Rotkäppchen", "Schneewittchen", "Froschkönig", "Wolf"]
```

Eine Möglichkeit zur Ausgabe jeder Figur aus der Liste, die wir bereits kennengelernt haben, wäre es, die Printfunktion mit dem Index von jedem Listeneintrag aufzurufen:


```python
print(figuren[0], figuren[1], figuren[2], figuren[3])
```

Dies ist aus drei Gründen nicht optimal:

- **Nicht skalierbar.** Stellen Sie sich vor, Sie müssen alle Elemente einer Liste ausgeben, die Hunderte von Elementen enthält.
- **Schwierig zu pflegen.** Wenn wir eine weitere Änderung hinzufügen wollen – zum Beispiel jeden String in Großbuchstaben umzuwandeln – müssten wir den Code für jedes Element in der Liste ändern, was nicht nachhaltig ist.
- **Fragil.** Die manuelle Nummerierung von Indexwerten für jedes Element in einer Liste kann dazu führen, dass Fehler weiter mitgeführt werden, wenn wir in diesem Schritt selbst einen Fehler machen.

Eine sogenannte **for-Schleife** ist die bessere Lösung. Sie wiederholt die `print()`-Anweisung für jeden `character`, also jede Einheit, in unserer Figuren-Liste.


```python
figuren = ["Rotkäppchen", "Schneewittchen", "Froschkönig", "Wolf"]

for character in figuren:
    print(character)
```

Beispielsweise wie bei unserem Regal. Ein Märchenerzähler geht durch unser Regal, nimmt sich jeweils das Element aus dem Regal und liest es vor bzw. gibt es aus: 

![for-schleife](../images/for_schleife_regal.png)

Ihm ist egal, wie viele Elemente in der Liste (im Regal) sind. Er geht von vorne bis hinten durch, bis er am Ende angekommen ist.

Wie für die anderen Variablen kann auch die Schleifenvariable einen beliebigen Namen von uns erhalten (hier "figur"). Im Gegensatz dazu muss die Syntax immer einem bestimmten Muster folgen: Am Ende der Zeile, mit der die Schleife beginnt, muss ein **Doppelpunkt** stehen und die nächste(n) Zeile(n) muss/müssen **eingerückt** sein. Alles, was nach dem Doppelpunkt eingerückt ist, gehört zur Schleife.

Schleifen sind robust im Umgang mit Datensammlungen wie bspw. Listen. Selbst wenn sich ein Wert in der Liste ändert, hinzu kommt oder entfernt wird, funktioniert die Schleife weiterhin. Für die Schleife ist es unerheblich, wie lang die Liste ist oder welche Werte darin enthalten sind.


Man sollte – wie bei allen Variablen – auch bei der Schleifenvariablen einen aussagekräftigen Namen wählen, damit Sie selbst sowie mögliche Nachnutzende den Code leichter verstehen können.

Man kann eine Schleife auch nutzen, um jedes Element eines Strings zu durchlaufen:

```python
for letter in "Der Froschkönig":
    print(letter)
```

## Bedingungen (if)

Doch wir können unsere Listeneinträge auch schon vor der Nutzung der for-Schleife auf Gegebenheiten, wie z. B. das Vorkommen von Zahlen, prüfen und nur über `print()` ausgeben lassen, wenn die Bedingung, die wir gesetzt haben, zutrifft.

Hierfür gibt es die sogenannten **if-Bedingungen**.

Ein einfaches Beispiel für eine if-Bedingung wäre:


```python
anzahl_maerchen = 86

if anzahl_maerchen > 50:
    print("Das ist eine große Märchensammlung!")
```

Wir sehen also, dass geprüft wird, ob die Variable *anzahl_maerchen* größer als der Wert 50 ist. Ist die Bedingung `True` (also wahr), wird "Das ist eine große Märchensammlung!" ausgegeben.

Doch was ist, wenn die *anzahl_maerchen* kleiner ist? Führen Sie folgenden Code aus:

```python
anzahl_maerchen = 10

if anzahl_maerchen > 50:
    print("Das ist eine große Märchensammlung!")
```

Da *anzahl_maerchen* in diesem Beispiel kleiner als 10 ist und die Bedingung damit nicht erfüllt ist, passiert erstmal gar nichts. Wir erhalten also keine Ausgabe. Hierbei handelt es sich nicht um einen Fehler, denn wir haben ja nicht bestimmt, was ausgegeben werden soll, wenn die Bedingung nicht erfüllt ist. Der Code wird in diesem Fall einfach ohne Ausgabe weiterlaufen.

Da dies für uns natürlich nicht hilfreich ist, kann noch ein **else** hinzugefügt werden, um der Bedingung die Information zu geben, was passieren soll, wenn die Bedingung nicht zutrifft:


```python
anzahl_maerchen = 10

if anzahl_maerchen > 50:
    print("Das ist eine große Märchensammlung!")
else:
    print("Das ist eine kleine Märchensammlung.")
```

Nun werden wir aber nicht nur auf Situationen treffen, die mit "entweder ... oder" gelöst werden können, sondern es müssen auch darüber hinaus Bedingungen geprüft werden. Hierfür bietet sich das **elif** an. Elif ist eine Art zwischen-if, das bei `True` die festgelegte Ausgabe nutzt, aber bei `False` weiter übersprungen wird. Für unser Beispiel mit der Anzahl der Märchen in unserer Sammlung heißt das:


```python
anzahl_maerchen = 86

if anzahl_maerchen > 200:
    print("Das ist eine riesige Märchensammlung!")
elif anzahl_maerchen > 50:
    print("Das ist eine große Märchensammlung!")
else:
    print("Das ist eine kleine Märchensammlung.")
```

Bei Bedingungen lässt sich zwischen den Operatoren kleiner (`<`), größer (`>`), gleich (`==`), kleiner gleich (`<=`) oder größer gleich (`>=`) wählen. Wichtig ist hier zu beachten, dass "gleich" mit zwei Gleichheitszeichen (`==`) geschrieben wird, da ein einfaches Gleichheitszeichen (`=`) dafür vorgesehen ist, einer Variablen einen Wert zuzuweisen.

Nun können wir auch unsere Liste mit einer for-Schleife durchgehen, jedes Element anschauen und die Bedingung prüfen, ob es sich um eine Zahl handelt. Auch dafür gibt es wieder eine vorgefertigte Funktion: `isdigit()`

Wir schauen also unsere *wort_liste* an und wenn dort ein Element eine Zahl ist, wird diese einmal ausgegeben und danach aus der Liste entfernt:


```python
for eintrag in wort_liste:
    if eintrag.isdigit():
        print("Ich habe folgende Zahl entfernt:", eintrag)
        wort_liste.remove(eintrag)
```

Nun haben wir eine Liste von allen Wörtern in unserem Märchen und diese beinhaltet keine Zahlen mehr. Über einen Print-Befehl können wir uns die gesamte Liste oder nur einzelne Elemente anzeigen lassen. Dabei fällt bereits auf, dass die Zahl am Anfang entfernt wurde:


```python
print(wort_liste[:20])  # zeigt die ersten 20 Elemente der Liste
```

Wenn wir uns noch einmal die häufigsten Wörter anzeigen lassen:


```python
from collections import Counter

haeufigkeiten = Counter(wort_liste)
print(haeufigkeiten.most_common(10))
```

Es fällt auf, dass viele der enthaltenen Wörter für unsere spätere Sentimentanalyse nicht relevant sind. Dies ist z. B. der Fall bei "und", "oder", "aber", "in" usw.

Deshalb kann man sich eine Stoppwortliste schreiben und diese mit den Wörtern aus dem Märchen in der Liste abgleichen. Die Stoppwörter können wir dann wiederum aus unserer Liste streichen, um nur noch die wichtigsten Wörter, die für unsere Analyse von Bedeutung sein könnten, zu behalten.

## Übung

**Aufgabe 1:** Speichern Sie in einer Variablen `stopwords` eine Liste von Stoppwörtern. Nutzen Sie die Funktion `append()`. Fügen Sie mindestens 5 Stoppwörter hinzu (z.B. "und", "oder", "der", "die", "das", "in", "ein", "es", "zu", "den", "war", "nicht", "auch", "noch", "aber", "mit", "von", "als", "ist", "so" ...).


```python
# Speichern Sie eine leere Liste in der Variablen "stopwords"
stopwords = []
```

```python
# Speichern Sie mithilfe der .append() mindestens 5 Stoppwörter in Ihrer Liste
stopwords.append("und")
stopwords.append("der")
stopwords.append("die")
stopwords.append("das")
stopwords.append("oder")
```

```python
# Überprüfung:
pruefe_05_aufgabe_1()
```

Nun müssen wir diese Liste mit unserer `wort_liste` aus dem Märchen abgleichen. Dafür kann eine for-Schleife genutzt werden und jedes Wort in unserer `wort_liste` wird mit der Stoppwortliste abgeglichen. Am Ende wird jedes Wort, welches **nicht** in der Stoppwortliste vorkommt, in eine neue Liste (`listeOhneStoppwörter`) gespeichert, die wir später auswerten könnten. Wichtig: Die neue Liste wird vorher leer angelegt, bevor die for-Schleife läuft. Innerhalb der for-Schleife wird diese leere Liste dann mit den gefundenen Wörtern, die nicht in der Stoppwortliste vorkommen, gefüllt.


```python
listeOhneStoppwörter = []

for listeneintrag in wort_liste:
    if listeneintrag not in stopwords:
        listeOhneStoppwörter.append(listeneintrag)

print("Gefilterte Liste:", listeOhneStoppwörter)
```

Nun können wir die Zählung noch einmal durchführen:


```python
haeufigkeitenOhneStoppwörter = Counter(listeOhneStoppwörter)
print(haeufigkeitenOhneStoppwörter)
```

Wir können uns auch über die Funktion `most_common()` die häufigsten gezählten Wörter anzeigen lassen:


```python
# Die 10 häufigsten Wörter ausgeben
print(haeufigkeitenOhneStoppwörter.most_common(10))
```

Aber wir sehen, dass immer noch einige Wörter dabei sind, die wir für unsere Sentimentanalyse nicht brauchen. Bevor wir die Lösung dafür, wie wir nur noch an die Wörter gelangen, die wir wirklich benötigen, erarbeiten, sollten wir anhand einer Abschlussprüfung sichergehen, dass wir die Grundlagen gut verstanden haben.
