## Variablen

Jeder Python-Interpreter kann als Taschenrechner benutzt werden:


````python
3 + 5 * 4
````

Das ist toll, aber nicht sehr interessant. Um mit Daten etwas wirklich Sinnvolles anfangen zu können, müssen wir ihren Wert einer Variablen zuweisen. In Python können wir einer Variablen einen Wert zuweisen, indem wir das Gleichheitszeichen (=) verwenden. 

Im nachfolgenden Codeschnipsel ordnet Python der Variablen _märchen_ die Zahl 86 zu und der Variablen _name_ den Wert 'Grimm' in einfachen Anführungszeichen. Namen, Wörter oder Sätze, die in einer Variablen gespeichert werden, müssen in einfachen oder doppelten Anführungszeichen geschrieben werden. Die Erläuterung dazu, wieso das so ist, erfolgt zu einem späteren Zeitpunkt in dieser Einheit.


````python
märchen = 86
name = 'Grimm'
````

Jetzt haben wir eine Art Speicher-Container, sprich die Variable _märchen_, und in der den Wert 86 gespeichert und wir haben in der Variablen _name_ den Wert 'Grimm' gespeichert. Im Grunde ist eine Variable also eine Art Speichercontainer für einen Wert. Und wir vergeben für diesen Container einen Namen.

Geben wir in einer Code-Zelle den Variablennamen ein und führen diese aus, so erscheint der Wert in der Ausgabe.


````python
märchen
````

Erwarteter Output: `86`


````python
name
````

Erwarteter Output: `'Grimm'`


---

### Hinweise

Variablennamen...

- können nicht mit einer Zahl starten.
- können keine Leerzeichen enthalten.
- können Unterstriche enthalten (typischerweise, um separate Wörter zu verbinden).
- sind case sensitive, beachten also die Groß- und Kleinschreibung. Das heißt: "name" und "Name" sind zwei unterschiedliche Variablen.

Für unser Fallbeispiel heißt das: 
- _märchen_ ist ein gültiger Variablenname. _0märchen_ ist dagegen kein gültiger Variablenname.
- _märchen_ and _Märchen_ wären zwei verschiedene Variablen.

---

Immer, wenn wir im Folgenden _märchen_ oder _name_ verwenden, wird Python den Wert nutzen, den wir der Variablen zugewiesen haben.


## Datentypen

Python kennt verschiedene Arten von Datentypen. Die drei gängigsten sind:
- Integer (Ganze Zahlen, abgekürzt: int)
- Floats (Kommazahlen, abgekürzt: float)
- Strings (Zeichenketten / Wörter, abgekürzt str). Insbesondere mit Strings werden wir nachfolgend verstärkt arbeiten.

In dem vorangegangenen Beispiel hat die Variable *märchen* den Datentyp Integer und die Variable *name* den Datentyp String.


### Eingebaute Python-Funktionen

Bevor wir uns intensiver mit Strings bzw. Zeichenketten beschäftigen, sollten wir noch einige grundlegende Informationen zu Integer und Floats anhand von eingebauten Python-Funktionen kennenlernen.

Sobald wir Daten mit Variablennamen gespeichert haben, können wir sie in Berechnungen verwenden.


````python
märchen = 86
märchen = 86 + 4
````

Um Aufgaben mit Daten und Variablen in Python auszuführen, stellt uns die Sprache mehrere eingebaute Funktionen zur Verfügung. Um Informationen auf dem Bildschirm anzuzeigen, benutzen wir die von Python zur Verfügung gestellte Funktion "print()":

*Hinweis:* Von Python bereitgestellte Funktionen/Funktionsnamen sind IMMER auf Englisch.


````python
print(märchen)
````

Erwarteter Output: `90`

Wir sehen also, dass *märchen* nun einen zugewiesenen Wert von 90 hat. Denn: durch unsere vorige Zuweisung (märchen = 86 + 4) erhält die Variable _märchen_ die Zahl 90. Das ist eine ganze Zahl und somit ein Integer.

Obwohl floats Gleitkommazahlen sind, werden sie in Python durch einen Punkt getrennt. Das probieren wir direkt aus: Wir weisen durch Ausführen des folgenden Codes der Variablen _float_ einen Wert von 10,4 zu:


````python
float = 10.4
````

Wenn wir eine ganze Zahl, also einen Integerwert, mit einem Floatwert addieren, erhalten wir dadurch wieder einen Floatwert, den wir in einer neuen Variablen namens _berechnung_float_ speichern:


````python
berechnung_float = märchen + float
print(berechnung_float)
````

Erwarteter Output: `100.4`

Aber: So wie integers und floats können wir auch Strings miteinander verbinden. Wir nehmen einen neuen String und 'addieren' den Variablennamen dazu:


````python
name = 'Grimm'
fullname = 'Gebrüder ' + name
print(fullname)
````

Erwarteter Output: `Gebrüder Grimm`

*Hinweis: Das Leerzeichen hinter 'Gebrüder' muss in die einfachen Anführungszeichen gesetzt werden, da sonst beide Strings ohne Leerzeichen direkt miteinander verbunden werden würden.*

Für den Fall, dass man mit Variablen arbeitet, bei denen man sich unsicher ist, welcher Datentyp sich hinter dem Wert verbirgt, gibt es in Python eine eingebaute Funktion, die uns den Datentyp ausgibt: type()


````python
print(type(berechnung_float),
      type(märchen),
      type(fullname),
      type(print))
````

Erwarteter Output: `<class 'float'> <class 'int'> <class 'str'> <class 'builtin_function_or_method'>`

- _berechnung_float_ ist ein Beispiel für eine Gleitkommazahl oder float. Dies sind Bruchzahlen.
- Der Wert der Variablen *märchen* ist 90, also eine ganze Zahl und somit ein integer.
- Die Variable *fullname* bezieht sich auf die Zeichenkette (str) von 'Gebrüder Grimm'.
- Die eingebaute Python-Funktion print() ist ebenfalls ein Objekt mit einem Typ, in diesem Fall ist es eine builtin_function_or_method.

Wenn wir eine Funktion verwenden, also einen "Funktionsaufruf" durchführen wollen, setzen wir den Namen der Funktion in Klammern. Die Klammern sind wichtig: Wenn sie weggelassen werden, wird die Funktion nicht richtig ausgeführt!

Wir können mit einem einzigen Aufruf von print() aber auch mehrere Dinge auf einmal anzeigen:


````python
print('Die', fullname, 'haben', märchen, 'Märchen geschrieben.')
````

Erwarteter Output: `Die Gebrüder Grimm haben 90 Märchen geschrieben.`

Um den Wert der Variablen *märchen* zu ändern, müssen wir *märchen* mit dem Gleichheitszeichen (=) einen neuen Wert zuweisen:


````python
märchen = 200
print(märchen)
````

Erwarteter Output: `200`

---

##### Tipp:
Eine Variable ist wie eine Klebenotiz auf dem ein Name steht: Wenn man einer Variablen einen Wert zuweist, ist das so, als würde man den Klebezettel auf einen Wert kleben.

Anhand dieser Analogie können wir untersuchen, wie die Zuweisung eines Wertes an eine Variable die Werte anderer, scheinbar verwandter Variablen nicht verändert.


````python
# Hinweis: Wir gehen für das Beispiel davon aus, dass die Gebrüder Grimm
# ca. das 2,8-fache an Sagen wie an Märchen geschrieben haben

sagen = 2.8 * märchen
print(sagen)
````

Erwarteter Output: `560.0`

Auf Basis der Anzahl der Märchen konnten wir uns also ausgeben lassen, wie viele Sagen die Gebrüder Grimm verfasst haben.

---

### Hinweis: 

Alles in einer Codezeile, was auf das Symbol "#" folgt, ist ein Kommentar, der von Python ignoriert wird. Kommentare ermöglichen es Programmierer:innen, Notizen oder Erklärungen für andere Programmierende oder aber für sich selbst zu hinterlassen, ohne, dass sie den Code beeinflussen.

---

Ändern wir nun die Anzahl der *märchen*:


````python
märchen = 240
print('Anzahl Märchen neu:', märchen, 'und Anzahl Sagen immer noch:', sagen)
````

Erwarteter Output: `Anzahl Märchen neu: 240 und Anzahl Sagen immer noch: 560.0`

Wir sehen: Die Anzahl der Märchen ist nun nicht mehr 90, sondern 240. Denn: Wenn Variablen neu zugewiesen werden, wird der alte Wert 'vergessen'. Nur noch der aktuellste spielt dann eine Rolle.
Da die Variable *sagen* sich nicht merkt, woher der Wert kommt, wird der Wert auch nicht automatisch aktualisiert, obwohl sich der Wert für die Variable *märchen* geändert hat.

**3.** Was bewirken die letzten drei Zeilen dieses Programms in einfachen Worten?


````python
x = 1.0
y = 3.0
swap = x
x = y
y = swap
````

*Lösung:*

```
swap = x  #  x = 1.0 y = 3.0 swap = 1.0
x = y     #  x = 3.0 y = 3.0 swap = 1.0
y = swap  #  x = 3.0 y = 1.0 swap = 1.0
```

Diese drei Zeilen tauschen die Werte in `x` und `y` aus, indem sie die Variable `swap` zur vorübergehenden Speicherung verwenden. Dies ist eine ziemlich verbreitete Praxis in Programmiersprachen. Nennt sich auch "Dreieckstausch".


**4.** Vorhersage von Werten

Wie lautet der Endwert der Variablen *position* im folgenden Programm? Versuchen Sie, den Wert vorherzusagen, ohne das Programm auszuführen, und überprüfen Sie dann Ihre Vorhersage.


````python
initial = "left"
position = initial
initial = "right"
print(position)
````

*Lösung:*

```
initial = "left"   # Initial is assigned the string "left"
```

Die letzte Zuordnung zur Position war „left".
