## Schritt 1: Was sagt Python?

Berechnen wir den Sentiment Score für "Hans im Glück" — so wie wir es gelernt haben:


```python
# Text laden und bereinigen
bereinigter_text_hans, wort_liste_hans = lade_bereinigten_text('HansGlueck.txt')

# Stoppwörter und Zahlen entfernen
gefiltert_hans = []
for wort in wort_liste_hans:
    if wort not in GERMAN_STOPWORDS and not wort.isdigit():
        gefiltert_hans.append(wort)

print('Wörter gesamt:', len(wort_liste_hans))
print('Wörter nach Filterung:', len(gefiltert_hans))
```

```python
# Sentiment Score berechnen
score_hans, details_hans = berechne_sentiment_mit_negation(gefiltert_hans)

print('Sentiment Score:', score_hans)
print()
```

```python
# Welche emotionalen Wörter wurden gefunden?
positive_hans = []
negative_hans = []

for wort, orig, negiert, end in details_hans:
    if end > 0:
        positive_hans.append(wort)
    elif end < 0:
        negative_hans.append(wort)

print('Positive Wörter (', len(positive_hans), '):', positive_hans)
print()
print('Negative Wörter (', len(negative_hans), '):', negative_hans)
```

### Ergebnis

Python sagt: "Hans im Glück" hat eine **positive Stimmung**. Es kommen Wörter wie "glück", "froh", "gold", "schön" vor. Laut unserer Analyse ist Hans ein fröhliches, positives Märchen.
